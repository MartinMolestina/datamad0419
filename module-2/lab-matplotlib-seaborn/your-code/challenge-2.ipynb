## Coding Session I - 30 minutes

From the Fitbit data, we want to visually understand: 

1. **How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?**
1. **Is Fitbitter more active on weekend or workdays?**

##### All plots must be in the same jupyter notebook cell.

Hints:

- Use `Months_encoded` and `Week_or Weekend` columns
- Use `matplolib.pyplot` object oriented API
- Set your size figure to `12,4`
- Explore `plt.sca`
- Explore `plt.xticks`
- Save your figures

```python
# Crear el dataframe con los datos necesarios
wkdy_stps = fitbit['Steps'].groupby(fitbit['Months_encoded']).mean()
wkdy_stps.head(5)

# Crear la figura y subplots necesarios
fig, axs = plt.subplots(figsize = (12,4), nrows = 1, ncols = 2)

# Plot del primer grafico
plt.sca(axs[0])
wkdy_stps.plot()
plt.ylabel('Avarage number of steps')
plt.xlabel('Months')
plt.title('Num of Steps vs. Months')
plt.xticks(list(range(13)), ['','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
          rotation =45, fontsize =12)

# Plot segundo grafico
wkdy_stps2 = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).mean().sort_values()
display (wkdy_stps2)
plt.sca(axs[1])
wkdy_stps2.plot(kind='bar', color = 'g', alpha=0.5)
plt.ylabel('Avarage number of steps')
plt.title('Average setps walked on day type')
plt.xticks(list(range(2)), ['Weekend','Workday'],rotation =0, fontsize =12)

#saving the figure
plt.savefig('steps.png')
```

## Coding Session II - 30 minutes

**Write a loop to plot 3 scatter plots of the following features:**

1. **`Minutes Lightly Active`  vs `Steps`**
1. **`Minutes Very Active` vs `Steps`**
1. **`Minutes Sedentary` vs `Steps`**

```python
col=['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary','Steps']
df=fitbit[col]
display(df.head())

fig,axes = plt.subplots(figsize = (12,20), nrows = 3, ncols = 1)
for i in range(3):
    plt.sca(axes[i])
    plt.scatter(df.iloc[:,i],df['Steps'], color = '#253E61', alpha=0.6)
    plt.xlabel(df.iloc[:,i].name)
    plt.ylabel('Steps')
```
